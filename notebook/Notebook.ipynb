```python
correlations = df.corr(numeric_only=True)['price'].sort_values()
print(correlations)
```

```python
import pandas as pd
from sklearn.feature_selection import f_regression
from sklearn.preprocessing import LabelEncoder

# Assuming your DataFrame is called df and your target is "price"
cat_features = ['CarName', 'fueltype', 'aspiration', 'doornumber', 'carbody',
                'drivewheel', 'enginelocation', 'enginetype', 
                'cylindernumber', 'fuelsystem']

# Encode categorical features temporarily
df_encoded = df.copy()
le = LabelEncoder()

for col in cat_features:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Apply ANOVA F-test (f_regression)
X_cat = df_encoded[cat_features]
y = df_encoded['price']

f_scores, p_values = f_regression(X_cat, y)

# Create a DataFrame with results
cat_scores = pd.DataFrame({
    'Feature': cat_features,
    'F-score': f_scores,
    'p-value': p_values
}).sort_values(by='F-score', ascending=True)

print("Categorical features sorted by F-score:\n", cat_scores)

# Remove weak features (set threshold F-score < 10, can be adjusted)
weak_cat_features = cat_scores[cat_scores['F-score'] < 10]['Feature'].tolist()
print("\nDropping weak categorical features:", weak_cat_features)

# Drop from original df
df.drop(columns=weak_cat_features, inplace=True)
```
